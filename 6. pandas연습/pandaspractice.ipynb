Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 

```python
df.Income_Category.value_counts()
```

```python
dic = {
    'Unknown' : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K' : 'b',
    '$60K - $80K' : 'c',
    '$80K - $120K' : 'd',
    '$120K +' : 'e',
}

df['newIncome'] = df.Income_Category.map(lambda x : dic[x])
df
```

Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

```python
def changeCat(x):
    if x == 'Unknown':
        return 'N'
    elif x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    else:
        return 'e'
df['newIncome'] = df.Income_Category.apply(changeCat)
```

```python
df
```

Customer_Age의 값을 이용하여 나이 구간을 AgeState 컬럼으로 정의하라. (0~9 : 0 , 10~19 :10 , 20~29 :20 … 각 구간의 빈도수를 출력하라

```python
df.loc[:,'AgeState'] = df['Customer_Age']//10
df['AgeState'].value_counts().sort_index()
```

```python
# 다른 방법
df['AgeState'] = df.Customer_Age.map(lambda x : x//10*10)
df['AgeState'].value_counts().sort_index()
```

Education_Level의 값중 Graduate단어가 포함되는 값은 1 그렇지 않은 경우에는 0으로 변경하여 newEduLevel 컬럼을 정의하고 빈도수를 출력하라

```python
# df['newEduLevel'] = df['Education_Level'].map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'] = df.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
df['newEduLevel'].value_counts()
```

Credit_Limit 컬럼값이 4500 이상인 경우 1 그외의 경우에는 모두 0으로 하는 newLimit 정의하라. newLimit 각 값들의 빈도수를 출력하라

```python
df['newLimit'] = df.Credit_Limit.map(lambda x : 1 if x>=4500 else 0)
df['newLimit'].value_counts()
```

Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

```python
def check(x):
    if x.Marital_Status == 'Married' and x.Card_Category == 'Platinum':
        return 1
    else:
        return 0
    

df['newState'] = df.apply(check, axis=1) # 행 방향으로 적용을 해줘야하기 때문에 axis=1을 해야한다.
df['newState'].value_counts()
```

Gender 컬럼값 M인 경우 male F인 경우 female로 값을 변경하여 Gender 컬럼에 새롭게 정의하라. 각 value의 빈도를 출력하라

```python
def check_gender(x):
    if x == 'M':
        return 'male'
    else:
        return 'female'
    

df['Gender'] = df.Gender.apply(check_gender) # 행 방향으로 적용을 해줘야하기 때문에 axis=1을 해야한다.
df['Gender'].value_counts()
```

년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다. 한버더 확인

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]
    if hour == '24':
        hour = '00:00:00'
        
        FinalDate = pd.to_datetime(date+ " "+hour) + datetime.timedelta(days=1) # 24시 같은 경우에는 날짜가 하나가 더해줘야하므로 timedelta를 사용함
        
    else:
        hour = hour + ':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    return FinalDate
df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.info()
```

```python
df
```

```python
# 다른방법
def date_ft(x):
    import datetime
    if x[-2:] != '24':
        x = x.replace(':', ' ')
        return pd.to_datetime(x)
    else:
        date = pd.to_datetime(x.split(':')[0]) + datetime.timedelta(days=1)
        return date
df['date'] = df['(년-월-일:시)'].apply(date_ft)
```

## 07_Merge , Concat

merge: 인덱스를 기준, 공통된 컬럼 기준으로 병합    
concat: 옆, 밑으로 다 붙일 수 있다.

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/mergeTEst.csv',index_col= 0)
```

```python
df1 = df.iloc[:4,:]
df2 = df.iloc[4:,:]

display(df1)
display(df2)
```

df1과 df2 데이터를 하나의 데이터 프레임으로 합쳐라.

```python
pd.concat([df1,df2])
```

```python
df3 = df.iloc[:2,:4]
df4 = df.iloc[5:,3:]

display(df3)
display(df4)
```

df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 둘다 포함하고 있는 년도에 대해서만 고려한다

```python
pd.concat([df3,df4], axis=0, join='inner')
```

df3과 df4 데이터를 하나의 데이터 프레임으로 합쳐라. 모든 컬럼을 포함하고, 결측치는 0으로 대체한다

```python
pd.concat([df3,df4],join='outer').fillna(0)
```

```python
df5 = df.T.iloc[:7,:3]
df6 = df.T.iloc[6:,2:5]

display(df5)
display(df6)
```

df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 두 데이터 모두 포함하는 데이터만 출력하라

```python
pd.merge(df5, df6,on='Algeria',how='inner')
```

df5과 df6 데이터를 하나의 데이터 프레임으로 merge함수를 이용하여 합쳐라. Algeria컬럼을 key로 하고 합집합으로 합쳐라

```python
pd.merge(df5,df6,on='Algeria',how='outer')
```
